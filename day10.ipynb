```python



class Bank:
  # public variable
  name='charan'
  # protected variable
  _number =123456789
  # private variable
  __balance= 1000

  def printName(self):
    print(self.name)
    print(self.__balance)

class Details(Bank):
  def _printNumber(self):
    print(self._number)
    print(self.__balance)

d = Details()
print(d._number)
d.printName()
d._printNumber()

b= Bank()
print(b.__balance)
```

```python
def division(n1:int, n2:int):
  try:
    return n1/n2
  except ZeroDivisionError as e:
    print(f"execption: {e}")
  except Exception as e:
    print(f"there is an execption {e}")

def multiply(n1:int, n2:int):
  return n1*n2

print(division(100,0))
print(division(100,'hello'))
print(multiply(10,20))
```
